Prime Number 



```python
def prime_num(n):
    if n==1 or n==0:
        return False
    for i in range(2,n):
        if n%i==0:
            return False
    return True


def all_prime_num(m):
    for i in range(2,m+1):
        if prime_num(i):
            print(i,end=" ")


if __name__=="__main__":
    print(prime_num(7))
    all_prime_num(100)
    
    
```
